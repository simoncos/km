```python
import pandas as pd

import email
import smtplib
import datetime
```

```python
# need extract json from .html first
df = pd.read_json('instapaper_highlights_export.json', encoding='utf8')
```

```python
df_today = df.sample(15)
```

```python
display(df_today)
```

```python
for i, ins in enumerate(df_today.to_dict(orient='records')):
    print(f'No.{i}')
    print(f'[Title]: {ins["title"]}')
    print(f'[Highlight]:\n{ins["highlight"]}')
    if ins['note']:
        print(f'[Note]:{ins["note"]}')
    print(f'[Source]: {ins["source"]}')
    print('-'*100)
```

```python
html_today = df_today.to_html()
date = datetime.datetime.now().strftime('%Y-%m-%d')
```

```python
headers = {
    'Content-Type': 'text/html; charset=utf-8',
    'Content-Disposition': 'inline',
    'Content-Transfer-Encoding': '8bit',
    'From': 'l.z.simon@hotmail.com',
    'To': 'l.z.simon@hotmail.com',
    'Date': date,
    'X-Mailer': 'python',
    'Subject': date + ' Instapaper Highlights'
}
msg = ''
for key, value in headers.items():
    msg += "%s: %s\n" % (key, value)    
msg += '\n' + html_today
```

```python
try:
    s = smtplib.SMTP("smtp.live.com", 587)
    # s.ehlo() # Hostname to send for this command defaults to the fully qualified domain name of the local host.
    s.starttls() #Puts connection to SMTP server in TLS mode
    # s.ehlo()
    s.login('l.z.simon@hotmail.com', input("Email Password: "))
    s.sendmail('l.z.simon@hotmail.com', 'l.z.simon@hotmail.com', msg.encode('utf-8'))
    s.quit()
    print('Sent email.')
except Exception as e:
    print('Cannot send email.')
    raise(e)
```
